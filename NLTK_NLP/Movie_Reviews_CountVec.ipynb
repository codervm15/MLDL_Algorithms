```python
documents = []

for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append(( movie_reviews.words(fileid) , category))
        
documents
```

```python
import random 

random.shuffle(documents)
```

## Data Cleaning

```python
from nltk.corpus import stopwords
import string
```

```python
punc = list(string.punctuation)

stop = stopwords.words('english')

stop = stop + punc

stop
```

```python
## Function to convert pos_tag to simple form

from nltk.corpus import wordnet

def get_simple_tag(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
## Function to remove stopwords and do the lemmatization

def clean_data(words):
    clean_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            word = lemmatizer.lemmatize(w, pos= get_simple_tag(pos[0][1]))
            clean_words.append(word.lower())
    return clean_words
```

```python
data = [ ( clean_data(doc) , category) for doc, category in documents]
```

```python
data[0]
```

## Converting data to sklearn classifier format

```python
## Sklearn classifiers requiers (X,Y) where X is features array, Y is target label array

categories = [category for doc, category in data]

categories
```

```python
text_doc = [" ".join(doc) for doc, category in data]
text_doc[0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(text_doc, categories)

```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer(max_features = 2000)

X_train_features = count_vec.fit_transform(X_train)
X_test_features = count_vec.transform(X_test)


count_vec.get_feature_names()
```

```python
## This will contain 1500 documents with 2000 features 
X_train_features
```

```python
X_test_features
```

## Classifiaction using SVM

```python
from sklearn.svm import SVC

from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf1 = SVC()
clf1.fit(X_train_features, Y_train)
```

```python
# Predictions on training data

Y_train_pred = clf1.predict(X_train_features)

print(confusion_matrix(Y_train, Y_train_pred))
print(classification_report(Y_train, Y_train_pred))
```

```python
# Predictions on testing data

Y_test_pred = clf1.predict(X_test_features)

print(confusion_matrix(Y_test, Y_test_pred))
print(classification_report(Y_test, Y_test_pred))
```
