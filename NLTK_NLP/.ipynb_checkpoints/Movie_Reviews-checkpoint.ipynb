## Data conversion to features & target labels

```python
## Here we create an array Documents which contain tuples of word list and corresponding category.
documents = []

for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
        
documents
```

```python
import random 
random.shuffle(documents)
documents
```

## Data Cleaning

```python
## Function to remove stop words and punctuations from word list.
from nltk.corpus import stopwords
import string

stops = stopwords.words('english')
punc = list(string.punctuation)

stops = stops + punc

stops
```

```python
## Function to convert pos tag to simple value, because lemmatizer needs simple value to do lemmatization.
from nltk.corpus import wordnet

def get_simple_tag(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# pos_mapping = {
#     'noun': 'n',
#     'verb': 'v',
#     'adjective': 'a',
#     'adverb': 'r',
# }
```

```python
## Function to clean the the word list by doing lemmatization

from nltk import pos_tag

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()


def clean_words(words):
    output = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos=get_simple_tag(pos[0][1]))
            #[0][1] because pos tag always return (word, tag) and we need only tag
            #clean_word = lemmatizer.lemmatize(w, pos=pos_mapping.get(pos[0][1], 'n'))
            output.append(clean_word.lower())
    return output
```

```python
documents  = [(clean_words(document), category) for document, category in documents]
```

## Building Feature Set

```python
training_document = documents[0:1500]
testing_document = documents[1500:]
```

```python
all_words = []
for doc in documents:
    all_words = all_words + doc[0]

```

```python
all_words
```

```python
len(all_words)
```

```python
import nltk
```

```python
## Here we want to take only 3000 most common occuring words as features
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)

common
```

```python
## Here we want to take only 3000 most common occuring words as features
## These 3000 features will be our features
features = []

for i in common:
    features.append(i[0])

features
```

```python
# def get_feature_dict(words):
#     current_features = {}
#     word_set = set(words)
#     for w in features:
#         current_features[w] = w in word_set
#     return current_features

def get_feature_dict(words):
    current_features = {}
    
    for w in features:
        if w in words:
            current_features[w] = True
        else:
            current_features[w] = False
    return current_features
```

```python
training_data = [ (get_feature_dict(doc) , category) for doc, category in training_document]
testing_data = [ (get_feature_dict(doc), category) for doc, category in testing_document]
```

```python
training_data[0]
```

## Classifiactio using NLTK Naive Bayes Classifier

```python
from nltk import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(clf, testing_data)
```

```python
clf.show_most_informative_features(15)
```
